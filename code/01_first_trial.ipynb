```python
%matplotlib inline

# Read in the image
image = cv2.imread('../data/03_hsv.png')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Convert from RGB to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# HSV channels
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value')
ax3.imshow(v, cmap='gray')
```

```python
# Define our color selection criteria in HSV values
lower_hue = np.array([25, 0, 20]) 
upper_hue = np.array([70, 255, 255])

# Now try HSV!

# Define the masked area in HSV space
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)

# mask the image
masked_image = np.copy(image)
masked_image[mask_hsv==0] = [0,0,0]

# Vizualize the mask
plt.imshow(masked_image)
```

```python
image = cv2.imread('../data/00_car_trial.png')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
result = image.copy()
# lower boundary RED color range values; Hue (0 - 10)
lower1 = np.array([0, 0, 0])
upper1 = np.array([10, 255, 100])

# upper boundary RED color range values; Hue (160 - 180)
lower2 = np.array([150,0, 0])
upper2 = np.array([179,255,100])
 
lower_mask = cv2.inRange(image, lower1, upper1)
upper_mask = cv2.inRange(image, lower2, upper2)
 
full_mask = lower_mask + upper_mask;
 
result = cv2.bitwise_and(result, result, mask=full_mask)
 
cv2.imshow('mask', full_mask)
cv2.imshow('result', result)
 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
